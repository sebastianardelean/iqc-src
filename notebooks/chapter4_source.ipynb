```python
# Python â‰¥3.5 is required
import sys
assert sys.version_info >= (3, 5)

# Common imports
import numpy as np
import os

# To plot pretty figures
#%matplotlib inline
# import matplotlib as mpl
import matplotlib.pyplot as plt
# mpl.rc('axes', labelsize=14)
# mpl.rc('xtick', labelsize=12)
# mpl.rc('ytick', labelsize=12)

# Where to save the figures
PROJECT_ROOT_DIR = "."
CHAPTER_ID = "chapter_4"
IMAGES_PATH = os.path.join(PROJECT_ROOT_DIR, "src_images", CHAPTER_ID)
os.makedirs(IMAGES_PATH, exist_ok=True)


def get_path(fig_name, fig_extension="pdf"):
    path = os.path.join(IMAGES_PATH, fig_name + "." + fig_extension)
    return path
```

```python
from qiskit import QuantumCircuit, transpile, QuantumRegister
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram
```

```python
qc1 = QuantumCircuit(2)
qc1.cz(0,1)
qc2 = QuantumCircuit(2)
# also a controlled-Z
qc2.h(1)
qc2.cx(0,1)
qc2.h(1)
```

```python
import matplotlib.pyplot as plt
```

```python
# Create a single figure with 2 subplots
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Draw the circuit directly in subplot (a)
qc1.draw(output="mpl", ax=axes[0])
axes[0].set_title("a)", y=-0.2)


qc2.draw(output="mpl", ax=axes[1])
axes[1].set_title("b)", y=-0.2)

plt.tight_layout()
plt.savefig(get_path("cnots"), bbox_inches="tight", pad_inches=0)
plt.show()
```

```python
qc1 = QuantumCircuit(2)
# a controlled-Y
qc1.sdg(1)
qc1.cx(0,1)
qc1.s(1)

```

```python
qc2 = QuantumCircuit(2)
# a controlled-H
qc2.ry(np.pi/4,1)
qc2.cx(0,1)
qc2.ry(-np.pi/4,1)

```

```python
# Create a single figure with 2 subplots
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Draw the circuit directly in subplot (a)
qc1.draw(output="mpl", ax=axes[0])
axes[0].set_title("a)", y=-0.2)


qc2.draw(output="mpl", ax=axes[1])
axes[1].set_title("b)", y=-0.2)

plt.tight_layout()
plt.savefig(get_path("cy_ch"), bbox_inches="tight", pad_inches=0)
plt.show()
```

```python
qc1 = QuantumCircuit(2)
# swaps states of qubits a and b
qc1.swap(0,1)
qc2 = QuantumCircuit(2)
# swap a 1 from a to b
qc2.cx(0,1) # copies 1 from a to b
qc2.cx(1,0) # uses the 1 on b to rotate the state of a to 0
```

```python
# Create a single figure with 2 subplots
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Draw the circuit directly in subplot (a)
qc1.draw(output="mpl", ax=axes[0])
axes[0].set_title("a)", y=-0.2)


qc2.draw(output="mpl", ax=axes[1])
axes[1].set_title("b)", y=-0.2)

plt.tight_layout()
plt.savefig(get_path("swap"), bbox_inches="tight", pad_inches=0)
plt.show()
```

```python
qr0 = QuantumRegister(1, name='c')
qr1 = QuantumRegister(1, name='t')
qc = QuantumCircuit(qr0,qr1)
theta = np.pi
qc.ry(theta/2, 1)
qc.cx(0,1)
qc.ry(-theta/2,1)
qc.cx(0,1)
fig=qc.draw(output="mpl")
fig.savefig(get_path("example_ry_rotation"), bbox_inches="tight", pad_inches=0)
display(fig)
```
