# Exercise 1

Using Python and Qiskit, create a quantum circuit with only one qubit $\vert 1 \rangle$. Using Qiskit Aer, simulate the circuit using the state vector method and print the state vector. Measure the qubit and rerun the simulation. This time, plot the histogram.

```python
qc = QuantumCircuit(1)
qc.x(0)
qc.save_statevector()

simulator = AerSimulator(method='statevector')
qc = transpile(qc, simulator)

# Run and get statevector
result = simulator.run(qc).result()
statevector = result.get_statevector(qc)
print(statevector)
```

```python
qc.measure_all()
simulator = AerSimulator(method='statevector')

qc = transpile(qc, simulator)
# Run and get counts
result = simulator.run(qc).result()
counts = result.get_counts(qc)
plot_histogram(counts)
```

# Exercise 6

```python
def not_gate(input_str):
    """ NOT gate.

    Parameters
        input_str (str): Input, encoded in qubit 0.

    Returns:
        QuantumCircuit: Output NOT circuit.
    """
    qc = QuantumCircuit(1)
    if input_str == '1':
        qc.x(0)
    qc.barrier()
    qc.x(0)
    return qc
```

```python
def xor_gate(input_a, input_b):
    """ XOR gate.

    Parameters
        input_a (str): Input, encoded in qubit 0.
        input_b (str): Input, encoded in qubit 1.

    Returns:
        QuantumCircuit: Output XOR circuit
    """
    qc = QuantumCircuit(2)
    if input_a == '1':
        qc.x(0)
    if input_b == '1':
        qc.x(1)
    qc.barrier()
    qc.cx(0,1)
    return qc
```

```python
def and_gate(input_a, input_b):
    qc = QuantumCircuit(3)
    if input_a == '1':
        qc.x(0)
    if input_b == '1':
        qc.x(1)
    qc.barrier()
    qc.ccx(0,1,2)
    return qc
```

```python
def nand_gate(input_a, input_b):
    qc = QuantumCircuit(3)
    if input_a == '1':
        qc.x(0)
    if input_b == '1':
        qc.x(1)
    qc.barrier()
    qc.ccx(0,1,2)
    qc.x(2)
    return qc
```

```python
def or_gate(input_a, input_b):
    qc = QuantumCircuit(3)
    if input_a == '1':
        qc.x(0)
    if input_b == '1':
        qc.x(1)
    qc.x(0)
    qc.x(1)
    qc.barrier()
    qc.ccx(0,1,2)
    qc.x(2)
    return qc
```

```python
def measure(qc):
    qc.measure_all()

    # Transpile for simulator
    simulator = AerSimulator()
    qc = transpile(qc, simulator)
    
    # Run and get counts
    result = simulator.run(qc, shots=1, memory=True).result()
    memory = result.get_memory(qc)
    return memory
```

```python
## Test NOT GATE
for inp in ['0', '1']:
    qc = not_gate(inp)
    out = measure(qc)
    print(f"NOT with input {inp} gives output {out}")
    fig = qc.draw(output = 'mpl')
    out_name = f"not_gate_{inp}"
    fig.savefig(
        get_path(out_name),
        bbox_inches='tight',
        pad_inches=0)
    display(fig)
    print('\n')
```

```python
## Test XOR Gate
for inpa in ['0', '1']:
    for inpb in ['0', '1']:
        qc = xor_gate(inpa,inpb)
        out = measure(qc)
        print(f"XOR with input {inpa} {inpb} gives output {out[0][0]}")
        fig = qc.draw(output = 'mpl')
        # out_name = f"not_gate_{inp}"
        # fig.savefig(
        #     get_path(out_name),
        #     bbox_inches='tight',
        #     pad_inches=0)
        display(fig)
        print('\n')
```

```python
## Test AND Gate
for inpa in ['0', '1']:
    for inpb in ['0', '1']:
        qc = and_gate(inpa,inpb)
        out = measure(qc)
        print(f"AND with input {inpa} {inpb} gives output {out[0][0]}")
        fig = qc.draw(output = 'mpl')
        # out_name = f"not_gate_{inp}"
        # fig.savefig(
        #     get_path(out_name),
        #     bbox_inches='tight',
        #     pad_inches=0)
        display(fig)
        print('\n')
```

```python
## Test NAND Gate
for inpa in ['0', '1']:
    for inpb in ['0', '1']:
        qc = nand_gate(inpa,inpb)
        out = measure(qc)
        print(f"NAND with input {inpa} {inpb} gives output {out[0][0]}")
        fig = qc.draw(output = 'mpl')
        # out_name = f"not_gate_{inp}"
        # fig.savefig(
        #     get_path(out_name),
        #     bbox_inches='tight',
        #     pad_inches=0)
        display(fig)
        print('\n')
```

```python
## Test OR Gate
for inpa in ['0', '1']:
    for inpb in ['0', '1']:
        qc = or_gate(inpa,inpb)
        out = measure(qc)
        print(f"OR with input {inpa} {inpb} gives output {out[0][0]}")
        fig = qc.draw(output = 'mpl')
        # out_name = f"not_gate_{inp}"
        # fig.savefig(
        #     get_path(out_name),
        #     bbox_inches='tight',
        #     pad_inches=0)
        display(fig)
        print('\n')
```

# Exercise 7

```python
qc = QuantumCircuit(4,2)
qc.x(0)
qc.x(1)
#encode inputs in qubit 0 and 1
qc.barrier()
qc.cx(0,2)
qc.cx(1,2)
qc.ccx(0,1,3)
qc.barrier()
qc.measure(2,0)
qc.measure(3,0)
```

```python
import matplotlib.pyplot as plt
```

```python
# Transpile for simulator
simulator = AerSimulator()
qc = transpile(qc, simulator)

# Run and get counts
result = simulator.run(qc, shots=1, memory=True).result()
counts = result.get_counts(qc)

# Create a single figure with 2 subplots
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Draw the circuit directly in subplot (a)
qc.draw(output="mpl", ax=axes[0])
axes[0].set_title("a)", y=-0.4)

# Plot histogram in subplot (b)
plot_histogram(counts, ax=axes[1])
axes[1].set_title("b)", y=-0.2)

plt.tight_layout()
plt.savefig(get_path("half_adder_both"), bbox_inches="tight", pad_inches=0)
plt.show()
```
